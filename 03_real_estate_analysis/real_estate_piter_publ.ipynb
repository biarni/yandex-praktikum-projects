# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Описание данных:

•	`airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)  
•	`balcony` — число балконов  
•	`ceiling_height` — высота потолков (м)  
•	`cityCenters_neares` — расстояние до центра города (м)  
•	`days_exposition` — сколько дней было размещено объявление (от публикации до снятия)  
•	`first_day_exposition` — дата публикации  
•	`floor` — этаж  
•	`floors_total` — всего этажей в доме  
•	`is_apartment` — апартаменты (булев тип)  
•	`kitchen_area` — площадь кухни в квадратных метрах (м²)  
•	`last_price` — цена на момент снятия с публикации  
•	`living_area` — жилая площадь в квадратных метрах (м²)  
•	`locality_name` — название населённого пункта  
•	`open_plan` — свободная планировка (булев тип)  
•	`parks_around3000` — число парков в радиусе 3 км  
•	`parks_nearest` — расстояние до ближайшего парка (м)  
•	`ponds_around3000` — число водоёмов в радиусе 3 км  
•	`ponds_nearest` — расстояние до ближайшего водоёма (м)  
•	`rooms` — число комнат  
•	`studio` — квартира-студия (булев тип)  
•	`total_area` — площадь квартиры в квадратных метрах (м²)  
•	`total_images` — число фотографий квартиры в объявлении

## Предобработка данных

### Столбцы и типы данных

```python
data = data.rename(columns={'cityCenters_nearest' : 'city_centers_nearest', 
                            'open_plan' : 'is_open_plan', 'studio' : 'is_studio'}) #переименуем столбцы
data.info() #проверим
```

```python
#изменим тип данных
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format="%Y/%m/%dT%H:%M:%S")
data.info()
```

```python
print(data['is_apartment'].unique())
display(data.head())
data.query('is_apartment == False')
```

`is_apartment` имеет пропуски и значения True и False. В описании данных сказано, что вся база - база квартир, тогда непонятно, почему часть значений False. Вероятно столбец нам не понадобится, поэтому корректировать его не будем.

### Дублирование в названиях насленных пунктов

```python
data['locality_name'].replace('ё', 'е', inplace=True, regex=True) #заменим ё на е в названиях населенных пунктов
data_copy = data #скопируем базу чтобы проверить неявные дубликаты из-за разного написания названия одного и того же населенного пункта
data_copy['locality_name'] = data_copy['locality_name'].str.lower() #приведем название населенного пункта к низкому регистру
print(data_copy['locality_name'].unique()) #проверим
data_copy.duplicated().sum() #проверим наличие неявных дубликатов
```

Неявные дубликаты не найдены

### Пропуски

```python
data['balcony'] = data['balcony'].fillna(0) #заполним пропуски в балконах нулями
data['balcony'].sort_values().unique() #проверим
```

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0) #заполним пропуски в парках нулями
data['parks_around3000'].sort_values().unique() #проверим
```

```python
data['ponds_around3000'] = data['ponds_around3000'].fillna(0) #заполним пропуски в прудах нулями
data['ponds_around3000'].sort_values().unique() #проверим
```

```python
import warnings
warnings.filterwarnings('ignore') #избавимся от предупреждений

#заполним медианами площадь кухни по населенным пунктам при условии, что квартира не свободной планировки и не студия
for i in data['locality_name'].unique():
    data.loc[(data['locality_name'] == i) 
             & (data['is_studio'] != True) 
             & (data['is_open_plan'] != True) 
             & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['locality_name'] == i), 'kitchen_area'].median()
data.info() #проверим, остались не заполнены 42 значений
```

```python
#заполним медианами жилую площадь по населенным пунктам
for i in data['locality_name'].unique():
    data.loc[(data['locality_name'] == i) 
             & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['locality_name'] == i), 'living_area'].median()
data.info() #проверим, остались не заполнены 22 значений
```

## Расчёты и добавление результатов в таблицу

```python
data['price_sq_meter'] = data['last_price'] / data['total_area'] #добавим цену за кв.м
data['ratio_living_to_total'] = data['living_area'] / data['total_area'] #добавим соотношение жилой площади к общей
data['ratio_kitchen_to_total'] = data['kitchen_area'] / data['total_area'] #добавим соотношение площади кухни к общей
data['weekday_of_start_exp'] = pd.DatetimeIndex(data['first_day_exposition']).weekday #добавим день недели публикации
data['month_of_start_exp'] = pd.DatetimeIndex(data['first_day_exposition']).month #добавим месяц публикации
data['year_of_start_exp'] = pd.DatetimeIndex(data['first_day_exposition']).year #добавим год публикации
data.head() #проверим
```

```python
def floor_category(data):
    """
    Возвращает категорию этажа: 1 -первый, 2 - другой, 3 - последний
    """
    
    if data['floor'] == 1:
        return 1
    if data['floor'] == data['floors_total']:
        return 3
    else:
        return 2

#применим функцию и создадим новый столбец с категориями этажей
data['floor_category'] = data.apply(floor_category, axis=1) 
data['floor_category'].unique() #проверим
```
